```python
from typing import Literal
import s3fs
fs = s3fs.S3FileSystem(anon=True)

def autoenum(region: Literal["EAST", "WEST", "HIMAWARI"]):
    options = {
        "EAST": "s3://noaa-goes16",
        "WEST": "s3://noaa-goes17",
        "HIMAWARI": "s3://noaa-himawari8",
    }
    print(f"class {region}(str, Enum):")
    for path in fs.ls(options[region]):
        if "." in path:
            continue
        member = path.split("/")[-1]
        print(f'\t{member.replace("-","_").upper()} = "{member}"')
```

```python
REQUEST = HIMAWARI.AHI_L2_FLDK_SST


def get_urls(prod: HIMAWARI | EAST | WEST, date: pd.Timestamp) -> list[str]:

    if isinstance(prod, HIMAWARI):
        baseurl = "s3://noaa-himawari8"
        globpattern = f"{date:%m}/{date:%d}/*/*L2P*.nc"
    elif isinstance(prod, EAST | WEST):
        baseurl = f"s3://noaa-goes{16 if EAST else 17}"
        globpattern = f"{date.dayofyear:03}/*/*.nc"
    else:
        raise Exception

    return fs.glob(f"{baseurl}/{prod}/{date:%Y}/{globpattern}")


urls = get_urls(REQUEST, pd.Timestamp.fromisoformat("2022-01-01"))

import xarray as xr

import io

from contextlib import contextmanager


@contextmanager
def open_mfdataset(urls: str):
    def generate():
        for url in urls:
            with fs.open(url, "rb") as f:
                yield io.BytesIO(f.read())

    memory = tuple(generate())
    try:
        yield memory
    finally:
        for buffer in memory:
            buffer.close()


with open_mfdataset(urls[:2]) as f:
    ds = xr.open_mfdataset(f, combine="nested", concat_dim="time", engine="h5netcdf")
    if isinstance(REQUEST, HIMAWARI):
        rename = {"ni": "lon", "nj": "lat"}
    else:
        rename = {"x": "lon", "y": "lat"}
        ds = ds.rename({"x": "lon", "y": "lat"})
    df = ds.rename(rename).reset_coords().to_dataframe()
```
